exercise 3.2

```python
signal = thinkdsp.Chirp(start=220, end=880)
wave1 = signal.make_wave(duration=3)
```

```python
wave1.segment(start=0, duration=0.014).plot()
```

```python
signal1 = thinkdsp.SawtoothSignal(200)
wave2 = signal1.make_wave(duration=3)
wave2.segment(start=0, duration=0.014).plot()
```

```python
Sawtoothchirp=wave1+wave2
Sawtoothchirp.segment(start=0, duration=0.014).plot()
```

```python
spectrum = Sawtoothchirp.make_spectrum()
spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

exercise 3.6

```python
def plot_spectrogram(wave, seg_length):
    spectrogram = wave.make_spectrogram(seg_length)
    print('Time resolution (s)', spectrogram.time_res)
    print('Frequency resolution (Hz)', spectrogram.freq_res)
    spectrogram.plot(high=700)
    thinkplot.config(xlabel='Time(s)', ylabel='Frequency (Hz)')
```

```python
wave5 = thinkdsp.read_wave('47290__pholonow__e.wav')
plot_spectrogram(wave5,512)
```

```python
wave6 = thinkdsp.read_wave('47292__pholonow__o.wav')
plot_spectrogram(wave6,512)
```

```python
wave7 = thinkdsp.read_wave('47294__pholonow__u.wav')
plot_spectrogram(wave7,512)
```
